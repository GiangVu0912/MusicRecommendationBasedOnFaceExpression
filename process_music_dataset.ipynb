```python
angry = df[df['mood']==1]
```

```python
angry = angry.groupby(['track_id','genre'])[['liked','calmness']].mean().sort_values(['liked'], ascending = False).reset_index()
```

```python
angry = angry[(angry['liked']>=0.5) &  (angry['calmness']>=0.5) ]
```

```python
sad = df[df['mood']==2]
```

```python
sad.head()
```

```python
sad = sad.groupby(['track_id','genre'])[['liked','tenderness']].mean().sort_values(['liked'], ascending = False).reset_index()
```

```python
sad = sad[(sad['liked']>=0.5) &  (sad['tenderness']>=0.5) ]
```

```python
neutral = df[df['mood']==3]
```

```python
neutral = neutral.groupby(['track_id','genre'])[['liked','joyful_activation']].mean().sort_values(['liked'], ascending = False).reset_index()
```

```python
neutral = neutral[(neutral['liked']>=0.5) &  (neutral['joyful_activation']>=0.5)]
```

```python
happy = df[df['mood']==5]
```

```python
happy = happy.groupby(['track_id','genre'])[['liked','disliked']].mean().sort_values(['liked'], ascending = False).reset_index()
```

```python
happy= happy[(happy['liked']>=0.7)]
```

```python
surprise = df[df['mood']==4]
```

```python
surprise  = surprise.groupby(['track_id','genre'])[['liked','power','tension']].mean().sort_values(['liked'], ascending = False).reset_index()
```

```python
surprise = surprise[(surprise['liked']>=0.5) &  (surprise['power']>0) & (surprise['tension']>0)]
```

```python
sad['emotion'] = 'sad'
surprise['emotion'] = 'surprise'
happy['emotion'] = 'happy'
neutral['emotion'] = 'neutral'
angry['emotion'] = 'angry'
```

```python
sad.drop(columns = ['liked','tenderness'], inplace = True)
sad
```

```python
surprise.drop(columns = ['liked', 'power', 'tension'], inplace = True)
surprise
```

```python
happy.drop(columns = ['liked','disliked'], inplace = True)
happy
```

```python
neutral.drop(columns = ['liked','joyful_activation'], inplace = True)
neutral
```

```python
angry.drop(columns = ['liked','calmness'], inplace = True)
angry
```

```python
emotion = pd.concat([angry,happy,neutral,sad,surprise],axis = 0)
```

```python
def new_trackid(row):
	if row['genre'] == 'classical':
		row['track_id'] = row['track_id']
	elif row['genre'] == 'electronic':
		row['track_id'] = row['track_id'] - 200
	elif row['genre'] == 'pop':
		row['track_id'] = row['track_id'] - 300
	else:
		row['track_id'] = row['track_id'] - 100
	return row['track_id']
```

```python
emotion['new_trackid'] = emotion.apply(new_trackid,axis =1)
```

```python
e = emotion.drop(columns = 'track_id')
e.rename(columns = {'new_trackid': 'track_id'}, inplace = True)
```

```python
e['track_id'] = e['track_id'].astype(str)
```

```python
song = pd.read_csv('song.csv')
```

```python
song.head()
```

```python
song['Genre'] = song['Genre'].apply(lambda x : x.lower())
```

```python
song['track_id'] = song['Name'].apply(lambda x : x.split('.')[0])
```

```python
s = song.drop(columns = ['Name'])
s = s.rename(columns = {'Genre': 'genre', 'Title': 'title', 'Contributing artists': 'artist'})
```

```python
full_emotion = pd.merge(e,s, how = 'left', on = ['track_id','genre'])
```

```python
all = pd.read_csv('all.csv')
```

```python
expression = pd.merge(full_emotion, all, how = 'left', on = 'title' )
```

```python
expression = expression[~expression['url'].isna()]
```

```python
expression.to_csv('expression.csv')
```
