```python
### IMPORTS ###
import re
import time
import pandas as pd

from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.keys import Keys
```

```python
from selenium.webdriver.common.by import By
```

```python
def start_driver(force_restart=False):
    global DRIVER
    
    if DRIVER is not None:
        if force_restart:
            close_driver()
        else:
            raise RuntimeError('ERROR: cannot overwrite an active driver. Please close the driver before restarting.')
    
    # Setting up the driver
    options = webdriver.ChromeOptions()
    options.add_argument('-headless') # we don't want a chrome browser opens, so it will run in the background
    options.add_argument('-no-sandbox')
    options.add_argument('-disable-dev-shm-usage')

    DRIVER = webdriver.Chrome('chromedriver',options=options)

# Wrapper to close driver if its created
def close_driver():
    global DRIVER
    if DRIVER is not None:
        DRIVER.close()
    DRIVER = None
```

```python
options = webdriver.ChromeOptions()
options.add_argument('-headless') # we don't want a chrome browser opens, so it will run in the background
options.add_argument('-no-sandbox')
options.add_argument('-disable-dev-shm-usage')

DRIVER = webdriver.Chrome('chromedriver',options=options)
```

```python
def get_product_info_single(i):
    d = {'title':'', 'url':''}
    try:
      name = i.text
      d['title'] = name
    except:
      d['title'] = False
    
    try:
      url = i.get_attribute('href')
      d['url'] = url
    except:
      d['url'] = False

    return d
```

```python
def get_product_info_from_page(page_url):
  global DRIVER
  data = []
  DRIVER.get(page_url) # Use the driver to get info from the product page
  time.sleep(5)
  
  elem = DRIVER.find_element_by_tag_name('body')
  no_of_pagedowns = 45
  while no_of_pagedowns:
      elem.send_keys(Keys.PAGE_DOWN)
      time.sleep(1)
      no_of_pagedowns-=1
  song   = DRIVER.find_elements_by_class_name('soundTitle__title')
  for i in song:
    song_dict = get_product_info_single(i)
    data.append(song_dict)
  return data

```

```python
df = pd.DataFrame(get_product_info_from_page('https://soundcloud.com/user77075486'))
```
